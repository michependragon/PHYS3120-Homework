### The semi-empirical mass formula revisted
As we saw in last week’s homework, the nuclear binding energy _B_ of an atomic nucleus with atomic number _Z_ and mass number A can be calculated to a good approximation using the semi-empirical mass formula:


\begin{equation}
B = a_1A - a_2A^{2/3} - a_3\frac{Z^2}{A^{1/3}} - a_4\frac{(A-2Z)^2}{A} + \frac{a_5}{\sqrt{A}}
\end{equation}

where (in units of millions of electron volts, MeV for short), the constants a1 = 15.67, a2 = 17.23, a3 = 0.75, a4 = 93.2, and 
\begin{equation}
a_5 = \text{0 if A is odd \\\\12.0  if A and Z are both even, \\\\-12.0 if A is even and Z is odd}
\end{equation}

and binding energy per nucleon _B/A_ for a given _A_ and _Z_.



__Part a__
 
Starting from that program, modify it so that it takes just a single value of the atomic number _Z_ and then goes through all values of _A = Z → 3Z_, to find the one that has the largest binding energy per nucleon. This is the most stable nucleus with the given atomic number. Have your program print out the value of _A_ for this most stable nucleus and the value of the binding energy per nucleon.

_Note:_
We expect the most stable nucleus to have _A_ in the vicinity of _2Z_ because most nuclei have about as many neutrons as protons, so going through the values of A from _Z_ to _3Z_ should be enough to find the most stable one.

```python


#from function that returns binding energy per nucleon
def bindingpernucleon(Z):
    #choosing a low number as the starting point for the highest binding energy per nucleon
    highestBdA = -1000
    #choosing a starting point for the most stable value of Z
    moststable = Z
    #loop through values from z to 3z
    for A in range(Z, 3*Z+1):
        #constants in MeV
        a1 = 15.67
        a2 = 17.23
        a3 = 0.75
        a4 = 93.2
        #if the remainder of A divided by 2 is 1, it is odd, and a5 is 0
        if A%2 == 1:
            a5 = 0
        #if the remainder of A/2 and Z/2 is 0, its even and a5 is 12
        elif A%2 == 0 and Z%2 == 0:
            a5 = 12
        #if the remainder of A/2 is 0 and Z/2 does not equal 0, a5 is -12
        elif A%2 == 0 and Z%2 != 0:
            a5 = -12
        #binding energy equation
        B = (a1*A-a2 * A**(2/3)-a3*(Z**2 / A**(1/3))-a4 * ((A - 2*Z)**2 / A) + a5 / A**(1/2))
        #BdA represents binding energy per nucleon
        BdA = B/A
        #if the binding energy per nucleon is greater than the current highest, make that number the highest
        if BdA > highestBdA:
            highestBdA = BdA
            #note the corresponding A value the most stable for that Z value
            moststable = A
    print("highest binding energy per nucleon is", str(highestBdA))
    print("most stable A value is", str(moststable))

Z = int(input("what is the atomic number, Z?: "))

bindingpernucleon(Z)
```

__Part c__
Modify your program further to calculate and plot the running average of the data,
defined by
\begin{equation}
Y_k = \frac{1}{2r+1} \sum_{m=-r}^{r} y_k+m
\end{equation}
where _r_ = 5 in this case (and the _yk_ values are the sunspot numbers). Have the program plot both the original data and the running average on the same graph, again over the range covered by the first 1000 data points.

```python

#creating a function for running average
def running_average(y, r=5):
    Y_k = np.zeros_like(y)  
    for k in range(len(y)):
        #the range for averaging
        start = max(0, k - r)  
        end = min(len(y) - 1, k + r) 
        #computing the average
        Y_k[k] = np.mean(y[start:end+1]) 
    #return sunspot numbers    
    return Y_k

runavg_sunspots = running_average(sun1000, r=5)

#plotting running avergae
plt.plot(month1000, runavg_sunspots, label="running average", color='pink')
#taking the first 1000 months
sun1000 = sunspots[:1001]
#redoing the same process as above
month1000 = sun1000[:,0]
sunnum1000 = sun1000[:, 1]
#plotting original data, first 1000 points
plt.plot(month1000,sunnum1000, label="original data", color='lightblue' )

#labelling x, y axis, and displaying legend
plt.xlabel("months since Jan 1749")
plt.ylabel("number of sunspots")
plt.title("sunspots and running average")
plt.legend()
```

__Part a__\
Make a plot of the so-called _deltoid_ curve, which is defined parametrically by the equations
\begin{equation}
x = 2cos θ + cos2θ 
\end{equation}
\begin{equation}
y = 2sinθ -sin2θ 
\end{equation}
where 0≤θ≤2π.Take a set of values of _θ_ between zero and _2π_ and calculate _x_ and _y_ for each from the equations above, then plot _y_ as a function of _x_.

```python
import numpy as np
import matplotlib.pyplot as plt

#taking a set of 50 values between 0 and two and pi, 
theta = np.linspace(0, 2 * np.pi, 50)

#using equations given to calculate x and y
x = 2*np.cos(theta)+np.cos(2*theta)
y = 2*np.sin(theta)-np.sin(2*theta)
#plotting
plt.plot(x, y, label="deltoid curve", color='lightgreen')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("deltoid curve")

```

__Part b__\
Taking the approach a step further, one can make a polar plot _r = f(θ)_ for some function f by calculate _r_ for a range of values of _θ_ and then converting _r_ and _θ_ to Cartesian coordinate using the standard equation _x = r_ cos θ, _y = r sinθ_. Use this method to make a plot of the Galilean spiral _r=θ^2_ for _0 ≤ θ ≤ 10π_.



```python

#generate 1000 values between 0 and 10 pi
theta = np.linspace(0, 10 * np.pi, 1000)
#r=theta^2
r = theta**2
#conversion to cartesian coordinates
x = r * np.cos(theta)
y = r * np.sin(theta)
#plotting the galilean spiral
plt.plot(x, y, label="Galilean Spiral", color="pink")
#format, labelling
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("galilean spiral")




```
